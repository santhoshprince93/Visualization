```python
import pandas as pd
Apples = pd.DataFrame(index=['Atiya', 'Abbas', 'Cornelia','Stephanie', 'Monte'],
                   data={'Apples':[20, 10, 40, 20, 50],'Oranges':[35, 40, 25, 19, 33]})
```

```python
Apples
```

```python
color=['.2',".7"]
ax = Apples.plot.bar(color=['.7','.3'], figsize=(16,4))
```

```python
ax = Apples.plot.kde(color=['.2',".7"], figsize=(10,4))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
fig.suptitle('Two Variable Plots')
Apples.plot.line(ax=ax1, title='Line plot')
Apples.plot.scatter(x='Apples', y='Oranges',ax=ax2, title='Scatterplot')
Apples.plot.bar(color=color, ax=ax3, title='Bar plot')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
fig.suptitle('One Variable Plots', size=20, y=1.02)
Apples.plot.kde(color=color, ax=ax1, title='KDE plot')
Apples.plot.box(ax=ax2, title='Boxplot')
Apples.plot.hist(color=color, ax=ax3, title='Histogram')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
Apples.sort_values('Apples').plot.line(x='Apples', y='Oranges',ax=ax1)
Apples.plot.bar(x='Apples', y='Oranges', ax=ax2,color='Orange')
Apples.plot.kde(x='Apples', ax=ax3)
```
